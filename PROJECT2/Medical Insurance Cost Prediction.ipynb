Now we have to remove the outliers for removing the skewness from the dataset

ZSCORE METHOD

```python
from scipy.stats import zscore

# to get the zscore
z=np.abs(zscore(df))
```

```python
z
```

```python
#checking the zscore values greater than 3
threshold=3
print(np.where(z>3))
```

```python
#randomly checking the zscore
z.loc[1317][2]
```

ZSCORE INDEXES PRODUCED ABOVE ARE MORE THAN 3, AS WE CHECKED ABOVE

```python
#checking the shape of the dataset before removing outliers
df.shape
```

REMOVING OUTLIERS FROM THE DATSET ON THE BASIS OF ZSCORE

```python
#Removing the outliers data
df1=df[(z<3).all(axis=1)]
```

```python
df1.shape
```

29 rows removed from the dataset

NOW CHECK THE IQR METHOD FOR REMOVAL OF OUTLIERS

```python
from scipy import stats
IQR = stats.iqr(df)
IQR
```

```python
Q1 = df.quantile(0.25)
Q3 =  df.quantile(0.75)
```

```python
#removing the outliers with IQR Method
df2 = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df2.shape
```

THERE ARE MORE THAN 600 rows of DATA WAS REMOVED IN IQR METHOD, SO WE WILL OPT FOR ZSCORE METHOD AS THERE IS ONLY FEW DATA LOSS
